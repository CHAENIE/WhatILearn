```python
# 데이터 분석 위한 pandas, 수치계산 위한 numpy
# 시각화 위한 seaborn, matplot.pyplot

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# jupyter notebook 있는 폴더에 data 폴더 만들어서 넣기
df=pd.read_csv("data/diabetes.csv")
df.shape
```

```python
df.head()
```

### 학습,예측 데이터셋 나누기

```python
# 데이터셋을 학습 세트과 예측 세트로 나눔
# df.shape는 튜플로 데이터를 가져오고, [0]이라는 인덱싱을 하면 앞자리인 df의 행의 갯수를 출력
# 여기서 나온 행의 갯수의 80%만 가져오는 연산을 하여 int() 함수를 이용해 정수로 나옴
# split_count 변수에 담아주고, 가장 마지막 줄을 출력하기 위해 split_count를 입력
split_count = int(df.shape[0] * 0.8)
split_count
```

```python
# ' : (콜론)'으로 슬라이싱이 가능
# 0번째 행부터 split_count번째 전까지 슬라이싱 한 행들을 명시적으로 복사하여 train에 넣기
# train data 80%
train = df[:split_count].copy()
train.shape
```

```python
# split_count부터 끝 행까지 슬라이싱
# test data 20%
test = df[split_count:].copy()
train.shape
```

### 학습, 예측에 사용할 컬럼

```python
feature_names=train.columns[:-1].tolist()
feature_names
```

### 정답값이자 예측해야 될 값(컬럼)

```python
label_name=train.columns[-1]
label_name
```

### 학습, 예측 데이터셋 만들기

```python
# 학습 세트 만들기 ex) 시험의 기출문제
# 행렬 형태
X_train=train[feature_names]
print(X_train.shape)
X_train.head()
```

```python
# 정답 값 만들기 ex) 기출문제의 정답
# 벡터 형태
y_train=train[label_name]
print(y_train.shape)
y_train.head()
```

```python
# 예측에 사용할 데이터셋 만들기. X_train의 열 개수가 같아야 함 ex) 실전 시험 문제
X_test=test[feature_names]
print(X_test.shape)
X_test.head()
```

```python
# 예측의 정답값 ex) 실전 시험 문제의 정답
# 예측이 맞았는지 확인하는 답안지 만들기. 학습 잘 되었는지 확인
y_test=test[label_name]
print(y_test.shape)
y_test.head()
```

## 1.2.3 의사결정나무로 학습과 예측하기

```python
# 머신러닝 알고리즘 가져오기
# 분류 문제 (if 당뇨병 '수치' 예측 -> 회귀 문제)
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model
```

### 학습(훈련)

```python
# X_train인 모의고사 문제를 갖고, 답안지인 y_train으로 문제를 풀음
model.fit(X_train, y_train)
```

### 예측

```python
# X-test는 실전 시험문제, 정답 직접 예측해야 함
# 정답열 제외, 풀어야 할 문제열만 넣기
y_predict=model.predict(X_test)
y_predict[:5]
```

## 1.2.4 예측한 모델의 성능 측정하기

### 트리 알고리즘 분석하기

```python
from sklearn.tree import plot_tree

plot_tree(model, feature_names=feature_names)
```

```python
# 밑으로 내려갈수록 샘플의 수 작아짐
from sklearn.tree import plot_tree

plt.figure(figsize=(20,20))
tree=plot_tree(model, feature_names=feature_names,
              filled=True,
              fontsize=10)
```

```python
# 그래프 더 크게
# 오버피팅 방지 위해 가지 개수 제한
import graphviz
from sklearn.tree import export_graphviz

dot_tree=export_graphviz(model, feature_names=feature_names,
                        filled=True)
graphviz.Source(dot_tree)
```

```python
# 컬럼 순서대로 feature 중요도 확인
model.feature_importances_
```

```python
# seaborn으로 중요도 수치
sns.barplot(x=model.feature_importances_, y=feature_names)
```

### 정확도(Accuracy) 측정하기

```python
# 실제값 - 예측값 빼면 같은 값은 0으로 나옴
# 여기에 절대값(abs) 씌운 게 1인 값이 다르게 예측된 값
# 실제 정답과 예측한 정답의 차를 합한 값 -> diff_count
diff_count=abs(y_test-y_predict).sum()
diff_count
```

```python
# 전체에서 26%를 틀렸음(잘못 예측)
abs(y_test-y_predict).sum()/len(y_test)
```

```python
# 73% 예측도
print("예측도는 ",(len(y_test)-diff_count)/len(y_test)*100)
```

```python
# 사이킷런 accuracy 메트릭 이용 or 모델에 탑재된 score함수 사용하여 정확도 확인 가능
from sklearn.metrics import accuracy_score

print("사이킷런 accuracy 메트릭 이용->",accuracy_score(y_test, y_predict))

print("모델에 탑재된 score함수 사용->",model.score(X_test, y_test)*100)
```
